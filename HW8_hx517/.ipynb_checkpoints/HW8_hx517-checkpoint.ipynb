cencus data

```python
cv = pd.read_csv('https://data.cityofnewyork.us/api/views/kku6-nxdu/rows.csv?accessType=DOWNLOAD')
```

```python
cv.head()
```

So now we have the predicted 2016 asthma hospital discharge rate for some Zipcode areas

```python
nyc = gpd.GeoDataFrame.from_file('ZIP_CODE_040114.shp')
```

```python
nyc.head()
```

```python
nyc = nyc.merge(cv,left_on='ZIPCODE',right_on='JURISDICTION NAME',how='outer')
```

```python
fig = plt.figure(figsize=[13,13])
ax = fig.add_subplot(111)
nyc.plot(column='POPULATION', cmap='Reds',ax=ax,legend=True,alpha=1)
ax.set_title('NYC POPULATION by ZipCode in 2016',size=15)
ax.set_axis_off()
ax.text(1066926,124719,u'\u25B2 \nN ', ha='center', fontsize=30, family='Arial', rotation = 0)
ax.legend()
```
